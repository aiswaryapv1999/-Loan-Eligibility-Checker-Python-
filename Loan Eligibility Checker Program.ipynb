```python
# Loan Eligibility check program

Age = int(input("Enter your Age"))
print("Age=",Age)
Monthly_income = float(input("Enter your Monthly Income"))
print("Monthly income=",Monthly_income)
Credit_score = int(input("Enter your Credit score"))
print("Credit score=",Credit_score)
Employment_status = input("Enter your Employment status, (employed / unemployed):").lower().strip()
print("Employment_status=",Employment_status)

if Age<21 or Monthly_income < 20000 or Credit_score < 650 or Employment_status == "unemployed":
    print("Loan Rejected")

    if Age<21:
        print("Age criteria not met")
    if Monthly_income < 20000:
        print("Insufficient income")
    if Credit_score < 650:
        print("Low credit score")
    if Employment_status == "unemployed":
        print("Employment required")
        
elif Age >= 21 and Monthly_income >= 20000 and Credit_score >= 650 and Employment_status == "employed":
    print("Loan Approved")

    if Credit_score >=750:
        print("Loan type = Premium Loan")
    else:
         print("Loan type = Standard Loan ")
```

```python
#Loan Eligibility Process

Age = int(input("Enter your Age"))
Monthly_income = float(input("Enter your Monthly Income"))
Credit_score = int(input("Enter your Credit score"))
Employement_status = input("Enter your Employment status, (Employed / Unemployed):").lower().strip()

if Age<21:
    print("Loan Rejected: Age criteria not met")
elif Monthly_income < 20000:
    print("Loan Rejected: Insufficient income")
elif Credit_score < 650:
    print("Loan Rejected: Low credit score")
elif Employement_status == "Unemployed":
    print("Loan Rejected: Employment required")
else:
    print("Loan Approved")
    if Credit_score >=750:
        print("Loan type = Premium Loan")
    elif Credit_score< 750 and Credit_score>=650:
         print("Loan type = Standard Loan ")

# Loan type classification

```
